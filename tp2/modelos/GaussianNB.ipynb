```python
import pandas as pd
import numpy as np

import sklearn
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
train_df = pd.read_csv("train.csv", low_memory=False)
test_df = pd.read_csv("test.csv", low_memory=False)
data = pd.read_csv("data_set.csv", low_memory = False)
```

```python
train_df = train_df[['id','precio']]
```

```python
features = pd.merge(train_df, data, on='id', how='inner')
features = features.fillna(0)
```

```python
features.describe()
```

```python
labels = features['precio']
features = features.drop(['id','precio'], axis=1)
```

```python
gnb = GaussianNB()
y_pred = gnb.fit(features, labels)
```

```python
set_test = pd.merge(data, test_df[['id']],on = "id", how = "inner")
ids = set_test["id"]
set_test = set_test.drop(columns=['id'])
set_test = set_test.fillna(0)
```

```python
df_kaggle_gaussiannb = pd.DataFrame(y_pred.predict(set_test))
df_kaggle_gaussiannb.head()
```

```python
df_kaggle_gaussiannb["id"] = ids
df_kaggle_gaussiannb.rename(columns = {0: 'target'},inplace = True) 
df_kaggle_gaussiannb = df_kaggle_gaussiannb[['id','target']]
```

```python
df_kaggle_gaussiannb.to_csv('ResultadosGNaiveBayes.csv',index=False)
```
