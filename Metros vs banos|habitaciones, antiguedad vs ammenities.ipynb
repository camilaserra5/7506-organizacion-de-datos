```python
# Importamos librerías de análisis de datos
import numpy as np
import pandas as pd

# plots
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Cargamos el dataframe
df = pd.read_csv('/home/gaston/Descargas/train.csv', index_col='id', parse_dates=['fecha'])
```

```python
df.info()
```

```python
# saco los nulos de habitaciones y metrostotales, me fijé los promedios sin sacar los nulos y llenando con 0 
# pero no vi ninguna diferencia en el promedio de los metros, se puede cambiar

dfcolumnas_habitaciones_nonulas = df.dropna(subset=['habitaciones'])
dfcolumnas_habitaciones_nonulas_metrosNoNulos = dfcolumnas_habitaciones_nonulas.dropna(subset=['metrostotales'])
dfagrupadas_habitaciones_nonulas_metrosNoNulos = df.groupby('habitaciones').\
agg({'metrostotales':'mean'}).sort_values(by = ['metrostotales'], ascending = False)

g = sns.lineplot(x=dfcolumnas_habitaciones_nonulas_metrosNoNulos.habitaciones,\
                 y=dfcolumnas_habitaciones_nonulas_metrosNoNulos.metrostotales,)
g.set_title("Cantidad de metros totales según la cantidad de habitaciones", fontsize=15)
g.set_xlabel("Cantidad de habitaciones", fontsize=12)
g.set_ylabel("Cantidad de metros totales", fontsize=12);
```

```python
dfcolumnas_habitaciones_nonulas = df.dropna(subset=['habitaciones'])
dfcolumnas_habitaciones_nonulas_metrosNoNulos = dfcolumnas_habitaciones_nonulas.dropna(subset=['metroscubiertos'])
dfagrupadas_habitaciones_nonulas_metrosNoNulos = df.groupby('habitaciones').\
agg({'metroscubiertos':'mean'}).sort_values(by = ['metroscubiertos'], ascending = False)

g = sns.lineplot(x=dfcolumnas_habitaciones_nonulas_metrosNoNulos.habitaciones,\
                 y=dfcolumnas_habitaciones_nonulas_metrosNoNulos.metrostotales,)
g.set_title("Cantidad de metros cubiertos según la cantidad de habitaciones", fontsize=15)
g.set_xlabel("Cantidad de habitaciones", fontsize=12)
g.set_ylabel("Cantidad de metros cubiertos", fontsize=12);
```

```python
dfcolumnas_banos_nonulas = df.dropna(subset=['banos'])
dfcolumnas_banos_nonulas_metrosNoNulos = dfcolumnas_banos_nonulas.dropna(subset=['metrostotales'])
dfagrupadas_habitaciones_nonulas_metrosNoNulos = df.groupby('habitaciones').\
agg({'metrostotales':'mean'}).sort_values(by = ['metrostotales'], ascending = False)

g = sns.lineplot(x=dfcolumnas_habitaciones_nonulas_metrosNoNulos.banos,\
                 y=dfcolumnas_habitaciones_nonulas_metrosNoNulos.metrostotales,)
g.set_title("Cantidad de metros totales según la cantidad de banos", fontsize=15)
g.set_xlabel("Cantidad de banos", fontsize=12)
g.set_ylabel("Cantidad de metros totales", fontsize=12);
```

```python
dfcolumnas_banos_nonulas = df.dropna(subset=['banos'])
dfcolumnas_banos_nonulas_metrosNoNulos = dfcolumnas_banos_nonulas.dropna(subset=['metroscubiertos'])
dfagrupadas_habitaciones_nonulas_metrosNoNulos = df.groupby('habitaciones').\
agg({'metroscubiertos':'mean'}).sort_values(by = ['metroscubiertos'], ascending = False)

g = sns.lineplot(x=dfcolumnas_habitaciones_nonulas_metrosNoNulos.banos,\
                 y=dfcolumnas_habitaciones_nonulas_metrosNoNulos.metrostotales,)
g.set_title("Cantidad de metros cubiertos según la cantidad de banos", fontsize=15)
g.set_xlabel("Cantidad de banos", fontsize=12)
g.set_ylabel("Cantidad de metros cubiertos", fontsize=12);
```

```python
dfcolumnas_antiguedad_nonulas = df.dropna(subset=['antiguedad'])
dfagrupadas_antiguedad_nonulas = df.groupby('antiguedad').agg({'piscina':'sum'}).sort_values(by = ['antiguedad'], ascending = True)
dfagrupadas_antiguedad_nonulas.reset_index(inplace=True)

g = sns.lineplot(x=dfagrupadas_antiguedad_nonulas.antiguedad, y=dfagrupadas_antiguedad_nonulas.piscina,)
g.set_title("Cantidad de propiedades con piscina según su antiguedad", fontsize=15)
g.set_xlabel("Antiguedad de la propiedad (años)", fontsize=12)
g.set_ylabel("Propiedades con piscina", fontsize=12);
```

```python
dfcolumnas_antiguedad_nonulas = df.dropna(subset=['antiguedad'])
dfagrupadas_antiguedad_nonulas = df.groupby('antiguedad').agg({'usosmultiples':'sum'}).sort_values(by = ['antiguedad'], ascending = True)
dfagrupadas_antiguedad_nonulas.reset_index(inplace=True)

g = sns.lineplot(x=dfagrupadas_antiguedad_nonulas.antiguedad, y=dfagrupadas_antiguedad_nonulas.usosmultiples,)
g.set_title("Cantidad de propiedades con SUM según su antiguedad", fontsize=15)
g.set_xlabel("Antiguedad de la propiedad (años)", fontsize=12)
g.set_ylabel("Propiedades con SUM", fontsize=12);
```

```python
dfcolumnas_antiguedad_nonulas = df.dropna(subset=['antiguedad'])
dfagrupadas_antiguedad_nonulas = df.groupby('antiguedad').agg({'gimnasio':'sum'}).sort_values(by = ['antiguedad'], ascending = True)
dfagrupadas_antiguedad_nonulas.reset_index(inplace=True)

g = sns.lineplot(x=dfagrupadas_antiguedad_nonulas.antiguedad, y=dfagrupadas_antiguedad_nonulas.gimnasio,)
g.set_title("Cantidad de propiedades con gimnasio según su antiguedad", fontsize=15)
g.set_xlabel("Antiguedad de la propiedad (años)", fontsize=12)
g.set_ylabel("Propiedades con gimnasio", fontsize=12);
```
