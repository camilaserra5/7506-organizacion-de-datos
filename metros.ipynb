```python
# Importamos librerías de análisis de datos
import numpy as np
import pandas as pd

# plots
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Cargamos el dataframe
df = pd.read_csv('train.csv', index_col='id', parse_dates=['fecha'])
```

```python
# nos quedamos con las columnas que nos interesan
df = df[['tipodepropiedad', 'metroscubiertos', 'metrostotales', 'precio']]
df.head(5)
```

```python
df[['metrostotales', 'tipodepropiedad']].sort_values(by = ['metrostotales'], ascending = False)
```

```python
grouped = df.groupby('tipodepropiedad').agg({'metrostotales':'mean', 'precio':'mean'})
grouped.sort_values(by = ['metrostotales'], ascending = False)
```
