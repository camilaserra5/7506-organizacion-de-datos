```python
# Importamos librerías de análisis de datos
import numpy as np
import pandas as pd

# plots
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
import datetime as dt
import time
```

```python
# Cargamos el dataframe
df = pd.read_csv('train.csv', index_col='id')
```

```python
# nos quedamos con las columnas que nos interesan
df = df[['titulo', 'tipodepropiedad','descripcion', 'antiguedad', 'fecha', 'precio', 'provincia', 'ciudad']]
df.head().sort_values(by = ['fecha'], ascending = False)
```

```python
df.isnull().any()
```

```python
df_fecha = df['fecha'].value_counts()
df_fecha
```

```python
fecha = df.fecha.apply(lambda x: dt.datetime.strptime(x,'%Y-%m-%d %H:%M:%S'))
```

```python
df['fecha'] = fecha
```

```python
anio = df.fecha.apply(lambda x: x.year)
df['anio'] = anio
hora = df.fecha.apply(lambda x: x.hour)
df['hora'] = hora
```

```python
hora.value_counts().head()
```

```python
plt.figure(figsize=(10, 4))
gr = sns.barplot(x=df.anio.value_counts().values , y=df.anio.value_counts().index ,  orient='h')
gr.set_title("Cantidad de publicaciones por anio", fontsize=15)
gr.set_xlabel("Cantidad de publicaciones", fontsize=12)
gr.set_ylabel("Ano", fontsize=12)
```

```python
precios2 = df['precio'].value_counts().sort_values(ascending=False).head(15).keys().tolist()
```

```python
precios2
```

```python
df_nuevo = df.loc[df['provincia'].isin(provincias)].loc[df['precio'].isin(precios)]\
    .groupby(['provincia','precio']).agg({'provincia':'count'})\
    .rename(columns = {'provincia':'cant_publ'})
```

```python
df_heatmap = df.loc[df['provincia'].isin(provincias)].loc[df['precio'].isin(precios2)]\
    .groupby(['provincia','precio']).agg({'provincia':'count'})\
    .rename(columns = {'provincia':'cant_publ'})
```

```python
df_nuevo
```

```python
df_heatmap = df_heatmap.reset_index()
pivot_post = df_heatmap.pivot_table(index= "precio", columns="provincia", values= "cant_publ")
pivot_post = pivot_post.sort_index(ascending=False)
plt.figure(figsize=(5,4.6))
cmap = sns.cm.rocket_r

g = sns.heatmap(pivot_post, cmap=cmap)
g.set_title("Distribución de Publicaciones")

fig = g.get_figure()
fig.set_figwidth(13)
fig.set_figheight(7)
```

```python
axs = df_nuevo.reset_index().pivot(index = 'provincia', columns = 'precio', values='cant_publ')\
    .plot(kind = 'bar', figsize =(12,8), fontsize = 12\
          ,title ='Cantidad de publicaciones segun provincia y precio')
axs.set_ylabel('Publicaciones', size = 15)
axs.set_xlabel('Costos según provincia', size = 15)
axs.legend(fontsize = 12)
axs.title.set_size(18)
```

```python
## veo que palabras son las más usadas en titulo y desc
```

```python
first_five = df.descripcion.str.split(n=5, expand=True)
first_five.head()
```

```python
words_df = pd.DataFrame(pd.concat([first_five[0],first_five[1],first_five[2],first_five[3],first_five[4]]))
words_df.columns = ['word']
words_df = words_df[(words_df['word']!='en') & (words_df['word']!='de') & (words_df['word']!='con')]
words_df.word = words_df.word.str.replace('<p>','')
wordcloud= WordCloud(width=1000, height=500, colormap="hsv", background_color='white')
wordcloud.generate_from_frequencies(frequencies=words_df.word.value_counts())

fig = plt.figure(figsize=(20,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.margins(x=0, y=0)
plt.axis('off')
plt.title('Frecuencias en primeras 5 palabras de las descripciones', fontsize=15)
plt.show()
```
