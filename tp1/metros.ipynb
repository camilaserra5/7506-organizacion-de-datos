```python
# Importamos librerías de análisis de datos
import numpy as np
import pandas as pd

# plots
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Cargamos el dataframe
df = pd.read_csv('train.csv', index_col='id', parse_dates=['fecha'])
```

```python
# nos quedamos con las columnas que nos interesan
df = df[['tipodepropiedad', 'metroscubiertos', 'metrostotales', 'precio']]
df.head(5)
```

```python
ax = df['metroscubiertos'].plot.hist(bins=20, figsize=(10,5))
ax.set_title("Metros cubiertos por publicación", fontsize=15)
ax.set_xlabel("Metros cubiertos", fontsize=12)
ax.set_ylabel("Cantidad de publicaciones", fontsize=12)
```

```python
ax = df['metrostotales'].plot.hist(bins=20, figsize=(10,5))
ax.set_title("Metros totales por publicación", fontsize=15)
ax.set_xlabel("Metros totales", fontsize=12)
ax.set_ylabel("Cantidad de publicaciones", fontsize=12)
```

```python
ax = df[['metroscubiertos','metrostotales']].plot.hist(bins=20, figsize=(10,5), alpha=0.8)
ax.set_title("Metros totales y cubiertos por publicación", fontsize=15)
ax.set_xlabel("Metros", fontsize=12)
ax.set_ylabel("Cantidad de publicaciones", fontsize=12)
```

```python
filtered = df[df['tipodepropiedad'].isin(df.tipodepropiedad.value_counts().head(3).index)]
filtered = filtered[['tipodepropiedad', 'metroscubiertos', 'metrostotales']]
```

```python
ax = filtered[filtered['tipodepropiedad'] == 'Apartamento'].dropna().plot.hist(alpha=0.8, bins=20, figsize=(10,5))
ax.set_title("Metros totales y cubiertos por publicación (Apartamento)", fontsize=15)
ax.set_xlabel("Metros", fontsize=12)
ax.set_ylabel("Cantidad de publicaciones", fontsize=12)
```

```python
ax = filtered[filtered['tipodepropiedad'] == 'Casa'].dropna().plot.hist(alpha=0.8, bins=20, figsize=(10,5))
ax.set_title("Metros totales y cubiertos por publicación (Casa)", fontsize=15)
ax.set_xlabel("Metros", fontsize=12)
ax.set_ylabel("Cantidad de publicaciones", fontsize=12)
```

```python
ax = filtered[filtered['tipodepropiedad'] == 'Casa en condominio'].dropna().plot.hist(alpha=0.8,bins=20, figsize=(10,5))
ax.set_title("Metros totales y cubiertos por publicación(Casa en condominio)", fontsize=15)
ax.set_xlabel("Metros", fontsize=12)
ax.set_ylabel("Cantidad de publicaciones", fontsize=12)
```
